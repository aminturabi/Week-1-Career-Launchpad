# WEEK 2 - DAY 1: DATA LOADING & EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the Dataset
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
# 2. Inspect Data Types
print(df.info())
```

```python

print(f"Dataset Shape: {df.shape}")
```

```python
# 3. Data Cleaning (The "TotalCharges" Trap)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print(f"\nMissing values in TotalCharges: {df['TotalCharges'].isnull().sum()}")
```

```python
# Drop these few missing rows
df.dropna(inplace=True)
```

```python
# 4. Target Variable Analysis (Churn)
# We need to know: Is the dataset balanced? (Do we have equal Yes vs No?)
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df, hue='Churn', palette='viridis', legend=False)
plt.title("Class Distribution: Churn vs Non-Churn")
plt.show()
```

```python
# Calculate exact percentage
churn_rate = (df['Churn'].value_counts(normalize=True) * 100)
print("\nChurn Rate:")
print(churn_rate)
```
